```python
x = [0,4,0,6,8,24,12,2]


def find_largest_num(list):
    largest = 0
    for i in range(len(list)):
      if largest > list[i]:
        largest = largest
      else:
        largest = list[i]

    return largest
print(find_largest_num(x))
```
